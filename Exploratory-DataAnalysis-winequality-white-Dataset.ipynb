```python
# checking the missing values
sns.heatmap(df.isnull(), cbar=False,yticklabels=False,cmap='viridis')
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(corr, cmap='Blues')
```

```python
# Correlation matrix
noOfVaraibleForHeatMap = 12
columns = df.corr().nlargest(noOfVaraibleForHeatMap, 'quality')['quality'].index
cm = df[columns].corr()
plt.figure9figsize=(10,6)
sns.heatmap(cm, annot=True, cmap = 'viridis')
```

```python
# let us check the outliers
columns_array = df.columns.values
number_of_columns=12
number_of_rows = len(columns_array)-1/number_of_columns
plt.figure(figsize=(number_of_columns,5*number_of_rows))
for i in range(0,len(columns_array)):
    plt.subplot(number_of_rows+1, number_of_columns,i+1)
    sns.set_style('whitegrid')
    sns.boxplot(df[columns_array[i]],color='green',orient='v')
    plt.tight_layout()
```

```python
# Let us check the skewness
plt.figure(figsize=(2*number_of_columns,5*number_of_rows))
for i in range(0,len(columns_array)):
    plt.subplot(number_of_rows+1, number_of_columns,i+1)
    sns.distplot(df[columns_array[i]],kde=True)
```
